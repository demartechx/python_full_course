```python
import pandas as pd
import numpy as np
df = pd.read_csv("Datas/weather.csv")
df
```

```python
new_df = df.replace([35,28], np.NaN) #replace values
new_df
```

```python
new_df = df.replace({
    'temperature': 32,
    'windspeed': 2,
    'event': 'Rain'
}, np.NaN) #specify columns and values to replace
new_df
```

```python
new_df = df.replace({
    35: np.NaN,
    'Snow': 'Sunny'
})
new_df
```

```python
new_df = df.replace('[A-Za-z]','',regex=True) #replace all strings with blank value
new_df
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'windspeed': '[A-Za-z]'
},'',regex=True) #replace all strings in selected columns with blank value
new_df
```
